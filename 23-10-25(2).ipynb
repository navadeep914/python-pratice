sep: Defines the delimiter used in the file (default is ,).
dtype: Specifies data types for each column to ensure consistent formatting.
skiprows: Skips the specified rows from the start or a list of row indices.
nrows: Reads only the specified number of rows from the file.
na_values: Specifies additional strings to treat as missing values (NaN

```python
import pandas as pd
df = pd.read_csv(
    "Uber_Drives_2016.csv",
    sep=",",                        # Use a comma as the delimiter
    dtype={"Age": "int64"},         # Ensure 'Age' column is of integer type
    skiprows=1,                     # Skip the first row
    nrows=5,                        # Read only the first 5 rows
    na_values=["NA", "Unknown"]     # Treat 'NA' and 'Unknown' as NaN
)
print(df)
```
